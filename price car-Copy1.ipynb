```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
```

```python
data = pd.read_csv('data_train.csv')
```

```python
data.head()
```

```python
data = data.drop('engine_fuel', axis=1)
```

```python
data.head()
```

```python
data = data.drop('number_of_photos', axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
engine_capacity_mean = data.engine_capacity.mean()
engine_capacity_mean
```

```python
data.engine_capacity = data.engine_capacity.fillna(engine_capacity_mean)
```

```python
data.isnull().sum()
```

```python
# Encode using LabelEncoder
# Encode 'manufacturer_name'
data.manufacturer_name = LabelEncoder().fit_transform(data.manufacturer_name)


# Encode 'transmission'
data.transmission = LabelEncoder().fit_transform(data.transmission)


# Encode 'color'
data.color = LabelEncoder().fit_transform(data.color)


# Encode 'engine_type'
data.engine_type = LabelEncoder().fit_transform(data.engine_type)


# Encode 'body_type'
data.body_type = LabelEncoder().fit_transform(data.body_type)


# Encode 'has_warranty'
data.has_warranty = LabelEncoder().fit_transform(data.has_warranty)


# Encode 'ownership'
data.ownership = LabelEncoder().fit_transform(data.ownership)


# Encode 'type_of_drive'
data.type_of_drive = LabelEncoder().fit_transform(data.type_of_drive)


# Encode 'is_exchangeable'
data.is_exchangeable = LabelEncoder().fit_transform(data.is_exchangeable)

data.head()
```

## Model Training

```python
model = Ridge()
```

```python
variables = data.drop(['price_usd'], axis=1)
target = data.price_usd

target.head()
```

```python
# Training
model.fit(variables, target)
```
